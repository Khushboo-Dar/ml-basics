```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import  train_test_split
time_studied=np.array([10,20,30,40,50,60]).reshape(-1,1)
scores=np.array([70,80,90,100,110,120]).reshape(-1,1)
time_train,time_test,score_train,score_test=train_test_split(time_studied,scores,test_size=0.3)
model=LinearRegression()
model.fit(time_train,score_train)
print(model.score(time_test,score_test))
plt.scatter(time_train,score_train)
plt.plot(np.linspace(0,70,100).reshape(-1,1),model.predict(np.linspace(0,70,100).reshape(-1,1)),'r')
plt.show()



```

1. LinearRegression: A ready-made tool from scikit-learn to perform linear regression 
2. train_test_split: A function to divide our data into two parts: training data and testing data.
3. reshape(-1, 1): Converts the 1D array into a 2D array because the model expects data in rows and columns. Here, the -1 automatically adjusts the number of rows.
5. test_size=0.3: Means 30% of the data will be for testing, and 70% for training.
6. score(): Calculates how well the model performs on the test data.
It gives a number called the R² score (ranges from 0 to 1). A higher number means the model fits the data better.
7. np.linspace(0, 70, 100): Creates 100 evenly spaced points between 0 and 70. This is used to draw the regression line.
